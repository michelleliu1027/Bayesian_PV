```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate
```

```python
address = 'data/degradation.xlsx'
x = []
y = []
df = pd.read_excel(address,sheet_name = 'normal data',usecols = [0],names = None,nrows = 5000)
df_li_Time = df.values.tolist()
df = pd.read_excel(address,sheet_name = 'normal data',usecols = [3],names = None,nrows = 5000)
df_li_L = df.values.tolist()
```

```python
x = np.array(df_li_Time).T
y = np.array(df_li_L).T 
x = x.reshape((5000))
y = y.reshape((5000))
tck = interpolate.splrep(x, y, s=0)
x_cubic = np.arange(0,5000,100)
y_cubic = interpolate.splev(x_cubic,tck, der=0)
```

```python
plt.figure()
plt.plot(x, y, 'x', x_cubic, y_cubic, '-')
plt.legend(['data','cubic spline'])
plt.tick_params(axis='y', colors = 'white')
plt.tick_params(axis='x', colors = 'white')
plt.show()
```

```python
s = interpolate.InterpolatedUnivariateSpline(x, y)
x_uni = np.arange(0,5000,100)
y_uni = s(x_uni)
```

```python
plt.figure()
plt.plot(x, y, 'x',x_uni,y_uni, x_cubic, y_cubic,'b')
plt.legend(['data','univariateSpline','cubic spline'])
plt.tick_params(axis='y', colors = 'white')
plt.tick_params(axis='x', colors = 'white')
plt.show()
```
